```python
import pandas as pd
from tqdm import tqdm

# Load the dataset
df = pd.read_parquet("eval_dataset.parquet")

# Ensure progress bar for visibility
tqdm.pandas()
def generate_faiss_response(row):
    query = row['query']
    retriever_info = row['retriever_rank']

    try:
        # Use the texts as chunks (convert to list to avoid np array issues)
        chunks = list(retriever_info.get('texts', []))
        if len(chunks) == 0:
            return {"answer": "", "input_tokens": 0, "output_tokens": 0}

        print(f"[INFO] Query: {query[:50]}...")
        print(f"[INFO] Chunks: {chunks[:3]}...")  # Print first 3 chunks for brevity
        answer, input_tokens, output_tokens = get_response_from_llm(query, chunks)
        return {
            "answer": answer,
            "input_tokens": input_tokens,
            "output_tokens": output_tokens
        }
        return
    except Exception as e:
        print(f"[Error] Query: {query[:50]}... -> {e}")
        return {"answer": "", "input_tokens": 0, "output_tokens": 0}


# Apply function row by row
df['FAISS'] = df.progress_apply(generate_faiss_response, axis=1)

# Save the updated DataFrame
df.to_parquet("eval_dataset_with_faiss.parquet", index=False)
```

```python
# Models evaluation
import os
import numpy as np
import pandas as pd
from stable_baselines3 import PPO, A2C, DQN
from sb3_contrib.ppo_recurrent import RecurrentPPO

from RankThenStopEnv import RankThenStopEnv
from Functions.llm import cached_evaluation_from_llm as reward_fn

# ‚öôÔ∏è Load your dataset
df = pd.read_parquet("eval_dataset_with_faiss_V1.parquet")
df.head(2)
```

```python
# üëá Choose the algorithm to load
algo_name = "a2c" # "a2c", "dqn", "ppo", "recurrent_ppo"
# model_path = f"./ppo_checkpoints/ppo_model_6000_steps.zip"
model_path = f"./train/{algo_name}/{algo_name}_final_model"

# ‚öôÔ∏è Select the right model class and policy
model_classes = {
    "ppo": PPO,
    "a2c": A2C,
    "dqn": DQN,
    "recurrent_ppo": RecurrentPPO,
}
policy_type = "MultiInputLstmPolicy" if algo_name == "recurrent_ppo" else "MultiInputPolicy"

# ‚úÖ Initialize environment
env = RankThenStopEnv(df=df, reward_fn=reward_fn, inference_mode=True)

# ‚úÖ Load the trained model
ModelClass = model_classes[algo_name]
model = ModelClass.load(model_path, env=env)

def get_selected_chunks_list(model, env, df):
    selected_chunks_list = []

    for _ in range(len(df)):
        obs = env.reset()
        done = False
        selected_chunks = []
        last_action = None  # Track the last action
        print("done")
        while not done:
            action, _ = model.predict(obs, deterministic=True)

            if last_action is None or action != last_action:
                print("action", action)

            last_action = action
            obs, reward, done, info = env.step(action)

        # After stopping, collect selected chunks for the current query
        selected_chunks = [env.chunks[i] for i in env.selected_indices]
        selected_chunks_list.append(selected_chunks)

    return selected_chunks_list



# üëá Get the selected chunks from the model
selected_chunks_list = get_selected_chunks_list(model, env, df)

# ‚úÖ Add predictions to the DataFrame
df[algo_name + "_selected_chunks"] = selected_chunks_list


# Add predictions to the DataFrame
df[algo_name+"selected_chunks"] = selected_chunks_list
```
